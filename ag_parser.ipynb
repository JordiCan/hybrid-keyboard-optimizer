```python
import os
import matplotlib.pyplot as plt
import re
from matplotlib.ticker import FuncFormatter

# Formatter para mostrar n√∫meros con separador de miles
def format_thousands(x, pos):
    return f'{x:,.0f}'

formatter = FuncFormatter(format_thousands)

def parse_results(file_path):
    """Parse results file containing (generation, fitness) tuples"""
    generations = []
    fitness_values = []
    with open(file_path, 'r') as file:
        for line in file:
            line = line.strip()
            if line.startswith('('):
                generation, fitness = eval(line)
                generations.append(generation)
                fitness_values.append(fitness)
    return generations, fitness_values
```

```python
# Experiment 1 - Population Size: Moby Dick
plt.figure(figsize=(12, 6))

pop_sizes = ['1000', '10000', '100000', '1000000']
for pop_size in pop_sizes:
    file_path = f'results/ag_results/exp1_popsize_{pop_size}_moby_dick_cln.txt'
    gens, fits = parse_results(file_path)
    plt.plot(gens, fits, label=f'Pop Size: {pop_size}', linewidth=2)

plt.title('Experiment 1: Population Size - Moby Dick', fontsize=14, fontweight='bold')
plt.xlabel('Generation', fontsize=12)
plt.ylabel('Fitness Value', fontsize=12)
plt.legend()
plt.grid(True, alpha=0.3)

# Formato del eje Y con separador de miles
ax = plt.gca()
ax.yaxis.set_major_formatter(formatter)

plt.tight_layout()
plt.show()
```

```python
# Parser de Resultados Finales - Todos los Experimentos
import pandas as pd
import re

def extract_final_fitness_and_layout(file_path):
    """Extract final fitness value and keyboard layout from results file"""
    with open(file_path, 'r') as f:
        content = f.read()
        lines = content.split('\n')
    
    # Get final fitness (last generation)
    final_fitness = None
    for line in reversed(lines):
        line = line.strip()
        if line.startswith('('):
            try:
                generation, fitness = eval(line)
                final_fitness = fitness
                break
            except:
                continue
    
    # Get keyboard layout - buscar el patr√≥n de lista con letras
    layout = None
    # Buscar patr√≥n como ['q', 'b', 'c', ...]
    layout_match = re.search(r'\[([^\]]+)\]', content)
    if layout_match:
        layout_str = '[' + layout_match.group(1) + ']'
        try:
            layout = eval(layout_str)
            # Convertir a string limpio
            layout = str(layout)
        except:
            layout = layout_str
    
    return final_fitness, layout

print("=" * 120)
print("RESULTADOS FINALES - ORDENADOS POR FITNESS (MEJOR A PEOR)")
print("=" * 120)

# EXPERIMENT 1: Population Size
print("\n" + "‚îÄ" * 120)
print("EXPERIMENT 1: POPULATION SIZE")
print("‚îÄ" * 120)
exp1_results = []
pop_sizes = ['1000', '10000', '100000', '1000000']

for pop_size in pop_sizes:
    for text, text_name in [('moby_dick_cln', 'Moby Dick'), ('wonderful_wizard_oz_cln', 'Wizard of Oz')]:
        file_path = f'results/ag_results/exp1_popsize_{pop_size}_{text}.txt'
        fitness, layout = extract_final_fitness_and_layout(file_path)
        exp1_results.append({
            'Configuration': f'PopSize_{pop_size}',
            'Text': text_name,
            'Final Fitness': fitness,
            'Layout': layout if layout else 'N/A'
        })

df_exp1 = pd.DataFrame(exp1_results).sort_values('Final Fitness')
print("\nMoby Dick - Ordenado por Fitness:")
print(df_exp1[df_exp1['Text'] == 'Moby Dick'][['Configuration', 'Final Fitness', 'Layout']].to_string(index=False))
print("\nWizard of Oz - Ordenado por Fitness:")
print(df_exp1[df_exp1['Text'] == 'Wizard of Oz'][['Configuration', 'Final Fitness', 'Layout']].to_string(index=False))

# EXPERIMENT 2: Tournament Selection
print("\n" + "‚îÄ" * 120)
print("EXPERIMENT 2: TOURNAMENT SELECTION")
print("‚îÄ" * 120)
exp2_results = []
k_values = ['2', '3', '5', '7', '10']

for k in k_values:
    for text, text_name in [('moby_dick_cln', 'Moby Dick'), ('wonderful_wizard_oz_cln', 'Wizard of Oz')]:
        file_path = f'results/ag_results/exp2_tournament_k{k}_{text}.txt'
        fitness, layout = extract_final_fitness_and_layout(file_path)
        exp2_results.append({
            'Configuration': f'Tournament_k{k}',
            'Text': text_name,
            'Final Fitness': fitness,
            'Layout': layout if layout else 'N/A'
        })

df_exp2 = pd.DataFrame(exp2_results).sort_values('Final Fitness')
print("\nMoby Dick - Ordenado por Fitness:")
print(df_exp2[df_exp2['Text'] == 'Moby Dick'][['Configuration', 'Final Fitness', 'Layout']].to_string(index=False))
print("\nWizard of Oz - Ordenado por Fitness:")
print(df_exp2[df_exp2['Text'] == 'Wizard of Oz'][['Configuration', 'Final Fitness', 'Layout']].to_string(index=False))

# EXPERIMENT 3: Mutation Rate
print("\n" + "‚îÄ" * 120)
print("EXPERIMENT 3: MUTATION RATE")
print("‚îÄ" * 120)
exp3_results = []
mutation_rates = ['0.05', '0.1', '0.15', '0.2', '0.3', '0.5', '0.75']

for rate in mutation_rates:
    for text, text_name in [('moby_dick_cln', 'Moby Dick'), ('wonderful_wizard_oz_cln', 'Wizard of Oz')]:
        file_path = f'results/ag_results/exp3_mutation_{rate}_{text}.txt'
        fitness, layout = extract_final_fitness_and_layout(file_path)
        exp3_results.append({
            'Configuration': f'Mutation_{rate}',
            'Text': text_name,
            'Final Fitness': fitness,
            'Layout': layout if layout else 'N/A'
        })

df_exp3 = pd.DataFrame(exp3_results).sort_values('Final Fitness')
print("\nMoby Dick - Ordenado por Fitness:")
print(df_exp3[df_exp3['Text'] == 'Moby Dick'][['Configuration', 'Final Fitness', 'Layout']].to_string(index=False))
print("\nWizard of Oz - Ordenado por Fitness:")
print(df_exp3[df_exp3['Text'] == 'Wizard of Oz'][['Configuration', 'Final Fitness', 'Layout']].to_string(index=False))

# EXPERIMENT 4: Elite Percentage
print("\n" + "‚îÄ" * 120)
print("EXPERIMENT 4: ELITE PERCENTAGE")
print("‚îÄ" * 120)
exp4_results = []
elite_pcts = ['5pct', '10pct', '15pct', '20pct', '30pct', '50pct']

for pct in elite_pcts:
    for text, text_name in [('moby_dick_cln', 'Moby Dick'), ('wonderful_wizard_oz_cln', 'Wizard of Oz')]:
        file_path = f'results/ag_results/exp4_elite_{pct}_{text}.txt'
        fitness, layout = extract_final_fitness_and_layout(file_path)
        exp4_results.append({
            'Configuration': f'Elite_{pct}',
            'Text': text_name,
            'Final Fitness': fitness,
            'Layout': layout if layout else 'N/A'
        })

df_exp4 = pd.DataFrame(exp4_results).sort_values('Final Fitness')
print("\nMoby Dick - Ordenado por Fitness:")
print(df_exp4[df_exp4['Text'] == 'Moby Dick'][['Configuration', 'Final Fitness', 'Layout']].to_string(index=False))
print("\nWizard of Oz - Ordenado por Fitness:")
print(df_exp4[df_exp4['Text'] == 'Wizard of Oz'][['Configuration', 'Final Fitness', 'Layout']].to_string(index=False))

# RESUMEN GENERAL - TOP 10 MEJORES RESULTADOS
print("\n" + "=" * 120)
print("TOP 10 MEJORES RESULTADOS GENERALES (TODOS LOS EXPERIMENTOS)")
print("=" * 120)

all_results = pd.concat([df_exp1, df_exp2, df_exp3, df_exp4])
top10_moby = all_results[all_results['Text'] == 'Moby Dick'].sort_values('Final Fitness').head(10)
top10_wizard = all_results[all_results['Text'] == 'Wizard of Oz'].sort_values('Final Fitness').head(10)

print("\nTOP 10 - Moby Dick:")
print(top10_moby[['Configuration', 'Final Fitness', 'Layout']].to_string(index=False))

print("\nTOP 10 - Wizard of Oz:")
print(top10_wizard[['Configuration', 'Final Fitness', 'Layout']].to_string(index=False))

print("\n" + "=" * 120)

# Guardar resultados en CSV para uso en Simulated Annealing
all_results.to_csv('results/all_ga_results_sorted.csv', index=False)
print("\n‚úì Resultados guardados en: results/all_ga_results_sorted.csv")
```

```python
# An√°lisis de Mejores Par√°metros por Experimento y Combinaci√≥n √ìptima
import pandas as pd
import re

def extract_final_fitness_and_layout(file_path):
    """Extract final fitness value and keyboard layout from results file"""
    with open(file_path, 'r') as f:
        content = f.read()
        lines = content.split('\n')
    
    # Get final fitness (last generation)
    final_fitness = None
    for line in reversed(lines):
        line = line.strip()
        if line.startswith('('):
            try:
                generation, fitness = eval(line)
                final_fitness = fitness
                break
            except:
                continue
    
    # Get keyboard layout
    layout = None
    layout_match = re.search(r'\[([^\]]+)\]', content)
    if layout_match:
        layout_str = '[' + layout_match.group(1) + ']'
        try:
            layout = eval(layout_str)
            layout = str(layout)
        except:
            layout = layout_str
    
    return final_fitness, layout

print("=" * 130)
print("AN√ÅLISIS DE MEJORES PAR√ÅMETROS POR EXPERIMENTO")
print("=" * 130)

# Diccionario para almacenar los mejores resultados
best_params = {
    'Moby Dick': {},
    'Wizard of Oz': {}
}

# EXPERIMENT 1: Population Size
print("\n" + "‚îÄ" * 130)
print("EXPERIMENT 1: POPULATION SIZE")
print("‚îÄ" * 130)

exp1_results = []
pop_sizes = ['1000', '10000', '100000', '1000000']

for pop_size in pop_sizes:
    for text, text_name in [('moby_dick_cln', 'Moby Dick'), ('wonderful_wizard_oz_cln', 'Wizard of Oz')]:
        file_path = f'results/ag_results/exp1_popsize_{pop_size}_{text}.txt'
        fitness, layout = extract_final_fitness_and_layout(file_path)
        exp1_results.append({
            'Parameter': pop_size,
            'Text': text_name,
            'Final Fitness': fitness,
            'Layout': layout
        })

df_exp1 = pd.DataFrame(exp1_results)
for text_name in ['Moby Dick', 'Wizard of Oz']:
    best_row = df_exp1[df_exp1['Text'] == text_name].sort_values('Final Fitness').iloc[0]
    best_params[text_name]['Population Size'] = {
        'value': best_row['Parameter'],
        'fitness': best_row['Final Fitness'],
        'layout': best_row['Layout']
    }
    print(f"\n{text_name}:")
    print(f"  Mejor Population Size: {best_row['Parameter']}")
    print(f"  Fitness: {best_row['Final Fitness']:,.2f}")
    print(f"  Layout: {best_row['Layout']}")

# EXPERIMENT 2: Tournament Selection
print("\n" + "‚îÄ" * 130)
print("EXPERIMENT 2: TOURNAMENT SELECTION")
print("‚îÄ" * 130)

exp2_results = []
k_values = ['2', '3', '5', '7', '10']

for k in k_values:
    for text, text_name in [('moby_dick_cln', 'Moby Dick'), ('wonderful_wizard_oz_cln', 'Wizard of Oz')]:
        file_path = f'results/ag_results/exp2_tournament_k{k}_{text}.txt'
        fitness, layout = extract_final_fitness_and_layout(file_path)
        exp2_results.append({
            'Parameter': k,
            'Text': text_name,
            'Final Fitness': fitness,
            'Layout': layout
        })

df_exp2 = pd.DataFrame(exp2_results)
for text_name in ['Moby Dick', 'Wizard of Oz']:
    best_row = df_exp2[df_exp2['Text'] == text_name].sort_values('Final Fitness').iloc[0]
    best_params[text_name]['Tournament k'] = {
        'value': best_row['Parameter'],
        'fitness': best_row['Final Fitness'],
        'layout': best_row['Layout']
    }
    print(f"\n{text_name}:")
    print(f"  Mejor Tournament k: {best_row['Parameter']}")
    print(f"  Fitness: {best_row['Final Fitness']:,.2f}")
    print(f"  Layout: {best_row['Layout']}")

# EXPERIMENT 3: Mutation Rate
print("\n" + "‚îÄ" * 130)
print("EXPERIMENT 3: MUTATION RATE")
print("‚îÄ" * 130)

exp3_results = []
mutation_rates = ['0.05', '0.1', '0.15', '0.2', '0.3', '0.5', '0.75']

for rate in mutation_rates:
    for text, text_name in [('moby_dick_cln', 'Moby Dick'), ('wonderful_wizard_oz_cln', 'Wizard of Oz')]:
        file_path = f'results/ag_results/exp3_mutation_{rate}_{text}.txt'
        fitness, layout = extract_final_fitness_and_layout(file_path)
        exp3_results.append({
            'Parameter': rate,
            'Text': text_name,
            'Final Fitness': fitness,
            'Layout': layout
        })

df_exp3 = pd.DataFrame(exp3_results)
for text_name in ['Moby Dick', 'Wizard of Oz']:
    best_row = df_exp3[df_exp3['Text'] == text_name].sort_values('Final Fitness').iloc[0]
    best_params[text_name]['Mutation Rate'] = {
        'value': best_row['Parameter'],
        'fitness': best_row['Final Fitness'],
        'layout': best_row['Layout']
    }
    print(f"\n{text_name}:")
    print(f"  Mejor Mutation Rate: {best_row['Parameter']}")
    print(f"  Fitness: {best_row['Final Fitness']:,.2f}")
    print(f"  Layout: {best_row['Layout']}")

# EXPERIMENT 4: Elite Percentage
print("\n" + "‚îÄ" * 130)
print("EXPERIMENT 4: ELITE PERCENTAGE")
print("‚îÄ" * 130)

exp4_results = []
elite_pcts = ['5pct', '10pct', '15pct', '20pct', '30pct', '50pct']

for pct in elite_pcts:
    for text, text_name in [('moby_dick_cln', 'Moby Dick'), ('wonderful_wizard_oz_cln', 'Wizard of Oz')]:
        file_path = f'results/ag_results/exp4_elite_{pct}_{text}.txt'
        fitness, layout = extract_final_fitness_and_layout(file_path)
        exp4_results.append({
            'Parameter': pct,
            'Text': text_name,
            'Final Fitness': fitness,
            'Layout': layout
        })

df_exp4 = pd.DataFrame(exp4_results)
for text_name in ['Moby Dick', 'Wizard of Oz']:
    best_row = df_exp4[df_exp4['Text'] == text_name].sort_values('Final Fitness').iloc[0]
    best_params[text_name]['Elite Percentage'] = {
        'value': best_row['Parameter'],
        'fitness': best_row['Final Fitness'],
        'layout': best_row['Layout']
    }
    print(f"\n{text_name}:")
    print(f"  Mejor Elite Percentage: {best_row['Parameter']}")
    print(f"  Fitness: {best_row['Final Fitness']:,.2f}")
    print(f"  Layout: {best_row['Layout']}")

# TABLA RESUMEN DE MEJORES PAR√ÅMETROS
print("\n" + "=" * 130)
print("TABLA RESUMEN: MEJORES PAR√ÅMETROS POR EXPERIMENTO")
print("=" * 130)

for text_name in ['Moby Dick', 'Wizard of Oz']:
    print(f"\n{'='*130}")
    print(f"  {text_name.upper()}")
    print(f"{'='*130}")
    
    summary_data = []
    for param_name, param_data in best_params[text_name].items():
        summary_data.append({
            'Experiment': param_name,
            'Best Value': param_data['value'],
            'Fitness': f"{param_data['fitness']:,.2f}"
        })
    
    df_summary = pd.DataFrame(summary_data)
    print(df_summary.to_string(index=False))

# COMBINACI√ìN √ìPTIMA RECOMENDADA
print("\n" + "=" * 130)
print("COMBINACI√ìN √ìPTIMA RECOMENDADA (Basada en mejores par√°metros individuales)")
print("=" * 130)

for text_name in ['Moby Dick', 'Wizard of Oz']:
    print(f"\n{'‚îÄ'*130}")
    print(f"  {text_name.upper()}")
    print(f"{'‚îÄ'*130}")
    
    optimal_config = {
        'Population Size': best_params[text_name]['Population Size']['value'],
        'Tournament k': best_params[text_name]['Tournament k']['value'],
        'Mutation Rate': best_params[text_name]['Mutation Rate']['value'],
        'Elite Percentage': best_params[text_name]['Elite Percentage']['value']
    }
    
    print("\n  Configuraci√≥n √ìptima:")
    for param, value in optimal_config.items():
        print(f"    ‚Ä¢ {param}: {value}")
    
    # Encontrar el mejor layout absoluto entre todos los experimentos
    all_experiments = [df_exp1, df_exp2, df_exp3, df_exp4]
    best_overall = None
    best_fitness = float('inf')
    
    for df in all_experiments:
        text_df = df[df['Text'] == text_name]
        if not text_df.empty:
            current_best = text_df.sort_values('Final Fitness').iloc[0]
            if current_best['Final Fitness'] < best_fitness:
                best_fitness = current_best['Final Fitness']
                best_overall = current_best['Layout']
    
    print(f"\n  Mejor Fitness Alcanzado: {best_fitness:,.2f}")
    print(f"  Mejor Layout: {best_overall}")

# Guardar configuraciones √≥ptimas en archivo
print("\n" + "=" * 130)
optimal_configs = []
for text_name in ['Moby Dick', 'Wizard of Oz']:
    optimal_configs.append({
        'Text': text_name,
        'Population_Size': best_params[text_name]['Population Size']['value'],
        'Tournament_k': best_params[text_name]['Tournament k']['value'],
        'Mutation_Rate': best_params[text_name]['Mutation Rate']['value'],
        'Elite_Percentage': best_params[text_name]['Elite Percentage']['value'],
        'Expected_Fitness': min(
            best_params[text_name]['Population Size']['fitness'],
            best_params[text_name]['Tournament k']['fitness'],
            best_params[text_name]['Mutation Rate']['fitness'],
            best_params[text_name]['Elite Percentage']['fitness']
        )
    })

df_optimal = pd.DataFrame(optimal_configs)
df_optimal.to_csv('results/optimal_ga_configuration.csv', index=False)
print("‚úì Configuraci√≥n √≥ptima guardada en: results/optimal_ga_configuration.csv")
print("=" * 130)
```
